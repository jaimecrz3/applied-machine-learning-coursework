Esto puede ser arriesgado si en dicho conjunto faltan patrones de algunas clases o si, en general, la muestra del conjunto de entrenamiento no está uniformemente distribuida (suele pasar con conjuntos de datos pequeños con frecuencia).

También después del entrenamiento, y como estamos ante un problema de clasificación, vamos a usar las siguientes métricas y la matriz de confusión:

* Precision: Tasa de aciertos entre los predichos como positivos, calculada como Precision = TP/(TP+FP)

* Recall: Tasa de aciertos entre los reales positivos, calculada como Recall = TP/(TP+FN)

* F1_score: Métrica que agrupa la precisión y el recall. Calculada como: F1 = 2*(Precision·Recall)/(Precision + Recall) = 2TP/(2TP + FP + FN)

donde:

* TP (True Positives). Elementos de la clase objetivo (1) que han sido bien clasificados.
* FP (False Positives). Elementos de la clase alternativa (0) que han sido clasificados en la clase objetivo (1).
* TN (True Negatives). Elementos de la clase alternativa (0) que han sido bien clasificados.
* FN (False Negatives). Elementos de la clase objetivo (1) que han sido clasificados en la clase alternativa (0).

MATRIZ DE CONFUSIÓN: Es una herramienta que permite visualizar el desempeño de un algoritmo de aprendizaje supervisado. Cada columna representa el número de predicciones de cada clase, mientras que cada fila representa a las instancias en la clase real. En la diagonal siempre van a estar los resultados correctos.

## **3.Preprocesado de datos**

Para este problema, al usar redes convolucionales, no necesitamos realizar ni Selección/Extracción de características ni reducción de la dimensionalidad, ya que el modelo tiene que aprender a partir de la información disponible.  

Por otro lado, En TensorFlow/Keras, las redes convolucionales esperan que las imágenes tengan una forma de 4 dimensiones: (n_samples, height, width, channels), donde:

* n_samples: número de imágenes

* height, width: dimensiones de cada imagen (28x28 en este dataset)

* channels: número de canales (1 para blanco y negro, 3 para RGB)

luego vamos a incluir una última dimensión a los datos de entrada.

```python
Train_images= np.expand_dims(Train_images, axis= -1)
Test_images= np.expand_dims(Test_images, axis= -1)

print('Tam. de entradas para training: {} y test: {}'.format(Train_images.shape, Test_images.shape))
```

Ya comentamos que ibamos a normalizar los datos, los vamos a forzar a que estén en el intervalo [0,1]:

```python
print('Rango inicial de los datos: [{}, {}]'.format(np.min(Train_images), np.max(Train_images)))

Train_images= Train_images/255
Test_images= Test_images/255
X_val = X_val/255

print('Rango final de los datos: [{}, {}]'.format(np.min(Train_images), np.max(Train_images)))
```

Por último, para mejorar la robustez de los datos, vamos a usar **DataAugmentation**, que consiste en generar variaciones de las imágenes mediante trasformaciones leves. Esto permite una generación de variaciones sintéticas para mejorar la generalización del modelo.

Primero creamos el generador de transformaciones datagen:
* rotation_range=10,  Esto rota aleatoriamente las imágenes entre -10 y 10 grados
* width_shift_range=0.1,   Esto desplaza horizontalmente hasta un 10% de la anchura
* height_shift_range=0.1,  Esto desplaza verticalmente hasta un 10% de la altura
* zoom_range=0.1,          Esto aplica zoom aleatorio dentro del 10%
* shear_range=0.1,        Esto aplica distorsión angular del 10%

Cuando despues llamamos a datagen.flow pasandole una imagen, lo que hace es que se aplica una combinación aleatoria de todas las transformaciones que se han definido en ImageDataGenerator.

Esto lo vamos a hacer de tal manera que por cada una de las 10 clases de ropa que disponemos se añadan 1000 imágenes nuevas, para que el dataset siga estando perfectamente balanceado.

```python
import numpy as np
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Generador de aumentación
datagen = ImageDataGenerator(
    rotation_range=10,
    width_shift_range=0.1,
    height_shift_range=0.1,
    zoom_range=0.1,
    shear_range=0.1
)

Train_images_augmented = []
Train_labels_augmented = []

for label in range(10):
    class_indices = np.where(Train_labels == label)[0]
    X_class = Train_images[class_indices]

    count = 0
    for i in range(len(X_class)):
        img = X_class[i].reshape(1, 28, 28, 1)
        aug_iter = datagen.flow(img, batch_size=1)

        for _ in range(5):  # Genera 5 aumentos por imagen
            aug_img = next(aug_iter)[0]
            Train_images_augmented.append(aug_img)
            Train_labels_augmented.append(label)
            count += 1
            if count >= 1000:
                break
        if count >= 1000:
            break

    print(f"Añadidas {count} imágenes aumentadas para clase {label}")

# Convertir a arrays y concatenar con el conjunto original
Train_images_augmented = np.array(Train_images_augmented)
Train_labels_augmented = np.array(Train_labels_augmented)

Train_images = np.concatenate([Train_images, Train_images_augmented], axis=0)
Train_labels = np.concatenate([Train_labels, Train_labels_augmented], axis=0)

print("Nuevo tamaño de X_train:", Train_images.shape)
print("Nuevo tamaño de y_train:", Train_labels.shape)
```

```python
print("Valor mínimo de píxeles:", Train_images.min())
print("Valor máximo de píxeles:", Train_images.max())
print("Media:", np.mean(Train_images))
print("Desviación estándar:", np.std(Train_images))
```

Como función de perdida saremos tf.keras.losses.sparse_categorical_crossentropy, que es la entropía cruzada categórica usada en problemas de clasificación multiclase cuando las etiquetas están codificadas como enteros (no en one-hot).

**¿Cómo funciona?**

Compara la probabilidad predicha por el modelo (y_pred) con la clase real (y_true), que es un entero y calcula la pérdida como el logaritmo negativo de la probabilidad asignada a la clase correcta.

Formula:

\begin{equation}
loss = -log(p_{clase\_correcta})
\end{equation}
Se promedia sobre todas las muestras del batch.

Como algoritmo de aprendizaje usaremos Adam, con una tasa de aprendizaje reducida. Es uno de los algoritmos más populares y recomendados para redes neuronales.

**¿Cómo funciona?**

Ajusta la tasa de aprendizaje de cada peso de forma individual, adaptativamente. Usa momentos del primer orden (la media del gradiente) y del segundo orden (la varianza).

El algoritmo Adam actualiza los parámetros del modelo usando las siguientes ecuaciones:

\begin{equation}
m_t = \beta_1 \cdot m_{t-1} + (1 - \beta_1) \cdot g_t
\end{equation}

\begin{equation}
v_t = \beta_2 \cdot v_{t-1} + (1 - \beta_2) \cdot g_t^2
\end{equation}

\begin{equation}
\hat{m}_t = \frac{m_t}{1 - \beta_1^t}
\end{equation}

\begin{equation}
\hat{v}_t = \frac{v_t}{1 - \beta_2^t}
\end{equation}

\begin{equation}
\theta_t = \theta_{t-1} - \alpha \cdot \frac{\hat{m}_t}{\sqrt{\hat{v}_t} + \epsilon}
\end{equation}

**Variables**

- g_t: Gradiente del paso actual.
- m_t: Promedio exponencial de gradientes pasados (momentum).
- v_t : Promedio exponencial de los cuadrados de gradientes (acelera/desacelera).
- beta_1, beta_2: Hiperparámetros de suavizado (usualmente 0.9 y 0.999).
- hat{m}_t, \hat{v}_t: Correcciones para evitar sesgos iniciales.
- alpha: Tasa de aprendizaje.
- epsilon: Pequeño número para evitar división por cero.
- theta_t: Parámetros (pesos) actualizados del modelo.

Adam acelera el aprendizaje al principio, pero se vuelve más cuidadoso cuando detecta ruido o incertidumbre.

* Tasa de aprendizaje del algoritmo de aprendizaje Adam. Por defecto vale 0.001.Controla qué tan grandes son los pasos que da el modelo en cada actualización de los pesos. Cuando previamente explicamos como funciona el algoritmo Adam y dimos las fórmulas, el valor alpha era la tasa de aprendizaje o learning rate.
* Vamos a probar con algoritmo de aprendizaje tanto con Adam como AdamW.
AdamW es una versión mejorada de Adam que corrige cómo se aplica la regularización L2. En Adam tradicional, la regularización L2 (también llamada weight decay) se mezcla incorrectamente con los momentos adaptativos, lo cual puede perjudicar el rendimiento en algunos casos.

Fórmula modificada de AdamW:

\begin{equation}
\theta_t = \theta_{t-1} - \alpha \cdot (\frac{\hat{m}_t}{\sqrt{\hat{v}_t} + \epsilon} + λ \cdot \theta_{t-1})
\end{equation}

Donde
$λ$ es el parámetro weight_decay. Esta penalización fuerza a los pesos a no crecer demasiado. Como se puede ver en la fórmula, si weight_decay = 0.0
entonces AdamW se comporta igual que Adam en la práctica.

Es recomendable usar AdanW cuando usas modelos más complejos o con muchas capas.
y cuando entrenas con muchas épocas y quieres mantener los pesos controlados.





Cuando se entrena un modelo con fit(), Keras guarda el estado del modelo por donde se ha quedado entrenando.
Esto quiere decir que si volvemos a usar fit(), el entrenamiento seguiría por
donde se ha quedado, y no empezaría desde el principio.

## **2.Definición del protocolo de validación experimental (entrenamiento, validación y test), junto con las métricas de evaluación del rendimiento que corresponda.**

Usando el dataset proporcionado, separado en training(60.000 imágenes) y test(10.000 imágenes), ahora simplemente tenemos que obtener el conjunto de validación.

Al contrario que en el caso del notebook de introduccion a Keras, si podemos usar el parámetro validation_split de la función fit.

Esto puede ser arriesgado si en dicho conjunto faltan patrones de algunas clases o si, en general, la muestra del conjunto de entrenamiento no está uniformemente distribuida (suele pasar con conjuntos de datos pequeños con frecuencia).

Si trabajasemos con un dataset más pequeño o desbalanceado, sí seria recomendable usar train_test_split(..., stratify=...) y guardar explícitamente la validación aparte.

A pesar de ello, como validation_split toma una partición del final del array, y no estamos seguros si puede haber muchas imágenes del mismo numero seguidas, vamos a usar:

```python
from sklearn.model_selection import train_test_split
XTrain2, X_val2, YTrain2, y_val2 = train_test_split(
    XTrain2, YTrain2, test_size=0.2, random_state=42, stratify=YTrain2
)
```

Respecto a las **métricas**, para entrenar el modelo vamos a usar 'MeanAbsoluteError (MAE)', esta métrica mide de forma directa cuánto difiere en promedio cada píxel reconstruido respecto al original y nos da una idea clara de cuánto error tiene, en media, cada píxel.

**¿Cómo funciona?**

La fórmula básica es:

\begin{equation}
\text{MAE} = \frac{1}{n} \sum_{i=1}^{n} \left| y_{\text{true}}^{(i)} - y_{\text{pred}}^{(i)} \right|
\end{equation}


Es decir: Calcula la diferencia absoluta entre el valor real y el valor predicho, píxel por píxel y luego hace la media de todas esas diferencias.

**¿Cómo se interpreta?**

* Una MAE de 0 significa que el modelo reconstruye la imagen perfectamente.

* Una MAE mayor indica que, en promedio, hay una diferencia más grande entre los píxeles originales y los reconstruidos.

```python
metrics2 = [tf.keras.metrics.MeanAbsoluteError()]
```

Por último, para mejorar la robustez de los datos, vamos a usar **DataAugmentation**, que consiste en generar variaciones de las imágenes mediante trasformaciones leves. Esto permite una generación de variaciones sintéticas para mejorar la generalización del modelo.

Primero creamos el generador de transformaciones datagen:
* rotation_range=10,  Esto rota aleatoriamente las imágenes entre -10 y 10 grados

Cuando despues llamamos a datagen.flow pasandole una imagen, lo que hace es que se aplica una combinación aleatoria de todas las transformaciones que se han definido en ImageDataGenerator.

Esto lo vamos a hacer de tal manera que por cada una de las 10 clases de ropa que disponemos se añadan 1000 imágenes nuevas, para que el dataset siga estando perfectamente balanceado.

```python
import numpy as np
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Generador de aumentación
datagen = ImageDataGenerator(
    rotation_range=10,
    width_shift_range=0.1,
    height_shift_range=0.1,
    zoom_range=0.1,
    shear_range=0.1
)

XTrain2_augmented = []
YTrain2_augmented = []

for label in range(10):
    class_indices = np.where(YTrain2 == label)[0]
    X_class = XTrain2[class_indices]

    count = 0
    for i in range(len(X_class)):
        img = X_class[i].reshape(1, 28, 28, 1)
        aug_iter = datagen.flow(img, batch_size=1)

        for _ in range(5):  # Genera 5 aumentos por imagen
            aug_img = next(aug_iter)[0]
            XTrain2_augmented.append(aug_img)
            YTrain2_augmented.append(label)
            count += 1
            if count >= 1000:
                break
        if count >= 1000:
            break

    print(f"Añadidas {count} imágenes aumentadas para clase {label}")

# Convertir a arrays y concatenar con el conjunto original
XTrain2_augmented = np.array(XTrain2_augmented)
YTrain2_augmented = np.array(YTrain2_augmented)

XTrain2 = np.concatenate([XTrain2, XTrain2_augmented], axis=0)
YTrain2 = np.concatenate([YTrain2, YTrain2_augmented], axis=0)

print("Nuevo tamaño de X_train:", XTrain2.shape)
print("Nuevo tamaño de y_train:", YTrain2.shape)
```

El Decoder recibe la representación comprimida generada por el Encoder (tamaño 7x7x16) y a través de varias capas de procesamiento, utiliza técnicas de upsampling y convolución para restaurar progresivamente la resolución de la imagen hasta llegar al tamaño original de 28x28x1.

En la primera capa, con UpSampling2D lo que se hace es pasar de tamaño de la imagen de entrada de 7x7 a 14x14. La operación de UpSampling hace que cada píxel de la imagen se repita, lo que es útil para agrandar la imagen. Luego aplicamos la capa convolucional con Relu como funcion de activación.

Reptimos este proceso para pasar a un tamaño de imagen de 28x28 y en la capa convolucional final rreducimos la cantidad de filtros a 1, generando una salida con un solo canal (en escala de grises).

```python
from tensorflow.keras.layers import UpSampling2D
def create_decoder():
  encoded_input = Input(shape=(7, 7, 16))

  x = UpSampling2D(size=(2, 2))(encoded_input)
  x = Conv2D(filters=32, kernel_size=(3, 3), padding='same')(x)
  x = ReLU()(x)

  x = UpSampling2D(size=(2, 2))(x)
  x = Conv2D(filters=64, kernel_size=(3, 3), padding='same')(x)
  x = ReLU()(x)

  x = Conv2D(filters=1, kernel_size=(3, 3), padding='same')(x)
  x = ReLU()(x)

  return Model(inputs=encoded_input, outputs=x)
```

**Autoencoder**

---

Un autoencoder es un tipo de red neuronal que tiene como objetivo aprender a representar datos de forma eficiente. Está compuesto por dos partes principales:

Encoder: Toma la entrada y la comprime en una representación de menor dimensión (embedding).

Decoder: Toma ese embedding y trata de reconstruir la entrada original lo mejor posible.

El objetivo es que el autoencoder aprenda a generar una representación comprimida (embedding) que contenga la información más importante sobre los datos, de forma que sea posible reconstruir la imagen a partir de esa representación.

El Autoencoder se entrena para minimizar la diferencia entre la imagen original y la imagen reconstruida. Es decir, el objetivo es que la pérdida entre la entrada y la salida (generalmente una función de error como el Mean Squared Error o MSE) sea lo más pequeña posible.

## **5. Selección y estimación de valores para hiperparámetros**

Nos piden que entrenmos el autoencoder un total de 50 épocas con tamaño de batch 1000 y usar una función de pérdida habitual para resolución de problemas de regresión. Vamos a ver que función de perdida es mejor para nuestro problema y también vamos a modificar los siguientes parámetros con el objetivo de discernir cual es el que obtiene mejores resultados.

* Tasa de aprendizaje del algoritmo de aprendizaje Adam. Por defecto vale 0.001.Controla qué tan grandes son los pasos que da el modelo en cada actualización de los pesos. Cuando previamente explicamos como funciona el algoritmo Adam y dimos las fórmulas, el valor alpha era la tasa de aprendizaje o learning rate.
* Vamos a probar con algoritmo de aprendizaje tanto con Adam como AdamW.
AdamW es una versión mejorada de Adam que corrige cómo se aplica la regularización L2. En Adam tradicional, la regularización L2 (también llamada weight decay) se mezcla incorrectamente con los momentos adaptativos, lo cual puede perjudicar el rendimiento en algunos casos.

Fórmula modificada de AdamW:

\begin{equation}
\theta_t = \theta_{t-1} - \alpha \cdot (\frac{\hat{m}_t}{\sqrt{\hat{v}_t} + \epsilon} + λ \cdot \theta_{t-1})
\end{equation}

Donde
$λ$ es el parámetro weight_decay. Esta penalización fuerza a los pesos a no crecer demasiado. Como se puede ver en la fórmula, si weight_decay = 0.0
entonces AdamW se comporta igual que Adam en la práctica.

Es recomendable usar AdanW cuando usas modelos más complejos o con muchas capas.
y cuando entrenas con muchas épocas y quieres mantener los pesos controlados.

Respecto a la

* Funcion de perdida: En un autoencoder, el objetivo es que la imagen de salida sea lo más parecida posible a la original. Es un problema de reconstrucción (regresión de pixeles), y algunas de las funciones de pérdida más habituales son:
 * MeanSquaredError (MSE): Esta función penaliza más fuertemente los errores grandes, ya que eleva al cuadrado la diferencia entre predicción y valor real.$y_i$ es el valor real, $\hat{y}_i$ es la predicción del modelo y $n$ es el número de muestras. Se usa cuandolos errores grandes son muy costosos o en casos donde los datos están limpios y sin mucho ruido.
  \begin{equation}
  \text{MSE} = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2
  \end{equation}
  * MeanAbsoluteError (MAE): Penaliza todos los errores de forma lineal, por lo que es más robusta frente a valores atípicos. Se suele usar en problemas con datos ruidosos o con outliers.
  \begin{equation}
  \text{MAE} = \frac{1}{n} \sum_{i=1}^{n} |y_i - \hat{y}_i|
  \end{equation}
 * Huber: Es una combinación entre MSE y MAE. Usa MSE para errores pequeños y MAE para errores grandes. Es útil cuando hay valores atípicos pero aún se quiere cierta sensibilidad a errores pequeños. El parámetro $\delta$ controla el punto de transición entre MSE y MAE.
\begin{equation}
L_\delta(y, \hat{y}) =
\begin{cases}
\frac{1}{2}(y - \hat{y})^2 & \text{si } |y - \hat{y}| \leq \delta \\
\delta \cdot (|y - \hat{y}| - \frac{1}{2}\delta) & \text{en otro caso}
\end{cases}
\end{equation}


En nuestro caso, como no tenemos datos ruidosos, vamos a trabajar con MSE y con Huber, para ver como funciona una combinación de la dos.



```python
from tensorflow.keras.optimizers import Adam, AdamW
from tensorflow.keras.models import Model
from tensorflow.keras import Input
configs = [
    {'optimizer': Adam(0.001), 'loss_function': 'mse'},
    {'optimizer': Adam(0.0005), 'loss_function': 'mse'},
    {'optimizer': Adam(0.001), 'loss_function': 'huber'},
    {'optimizer': Adam(0.0005), 'loss_function': 'huber'},
    {'optimizer': AdamW(0.001, weight_decay=1e-4), 'loss_function': 'mse'},
    {'optimizer': AdamW(0.001, weight_decay=1e-4), 'loss_function': 'huber'},
    {'optimizer': AdamW(0.0005, weight_decay=1e-4), 'loss_function': 'mse'},
    {'optimizer': AdamW(0.0005, weight_decay=1e-4), 'loss_function': 'huber'},
    {'optimizer': AdamW(0.001, weight_decay=1e-3), 'loss_function': 'mse'},
    {'optimizer': AdamW(0.001, weight_decay=1e-3), 'loss_function': 'huber'},
    {'optimizer': AdamW(0.0005, weight_decay=1e-3), 'loss_function': 'mse'},
    {'optimizer': AdamW(0.0005, weight_decay=1e-3), 'loss_function': 'huber'},
]

histories = []
val_maes = []

for i, config in enumerate(configs):
  print(f"Entrenando configuración {i+1}/{len(configs)} con {config}")

  encoder_ = create_encoder()
  decoder_ = create_decoder()

  # Entrada al autoencoder
  input_img = Input(shape=(28, 28, 1))
  encoded = encoder_(input_img)
  decoded = decoder_(encoded)

  # Modelo completo
  model = Model(inputs=input_img, outputs=decoded)
  model.compile(optimizer=config['optimizer'],
                loss=config['loss_function'],
                metrics=metrics2)

  history = model.fit(XTrain2, XTrain2,
                      validation_data=(X_val2, X_val2),
                      batch_size=1000,
                      epochs=20,
                      verbose=1)
  histories.append(history)
  val_maes.append(history.history['val_mean_absolute_error'][-1])

for i, (config, mae) in enumerate(zip(configs, val_maes)):
    print(f"Config {i}: Opt={config['optimizer'].__class__.__name__}, "
          f"Loss={config['loss_function']} → Val MAE: {mae:.4f}")


best_index = int(np.argmin(val_maes))
print(f"\n Mejor configuración: {best_index} con Val MAE = {val_maes[best_index]:.4f}")
print(f"Detalles: {configs[best_index]}")


plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(val_maes, marker='o')
plt.title("Validation MAE por configuración")
plt.xlabel("Configuración")
plt.ylabel("Val MAE")

plt.subplot(1, 2, 2)
for i, h in enumerate(histories):
    plt.plot(h.history['val_mean_absolute_error'], label=f'Config {i}')
plt.title("Curvas de validación MAE")
plt.xlabel("Época")
plt.ylabel("Val MAE")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
def sample_from_embedding(mean, std, num_samples=5):
    samples = []
    for _ in range(num_samples):
        noise = np.random.normal(loc=0.0, scale=1.0, size=std.shape)
        sample = mean + noise * std
        samples.append(sample)
    return np.array(samples)
```

```python
import matplotlib.pyplot as plt

num_samples = 5

plt.figure(figsize=(10, 20))
for digit in range(10):
    samples = sample_from_embedding(digit_embeddings_mean[digit],
                                    digit_embeddings_std[digit],
                                    num_samples=num_samples)


    decoded_imgs = decoder.predict(samples)

    for i in range(num_samples):
        plt.subplot(10, num_samples, digit * num_samples + i + 1)
        plt.imshow(decoded_imgs[i].squeeze(), cmap='gray')
        plt.axis('off')
        if i == 0:
            plt.ylabel(f'Dígito {digit}', fontsize=12)

plt.suptitle("Imágenes generadas por muestreo de embeddings por dígito", fontsize=16)
plt.tight_layout()
plt.show()
```

**¿Qué se obtiene?**

Para cada dígito, hemos generado nuevas imágenes artificiales que no estaban en el dataset original. Podemos ver que son visualmente similares al dígito correspondiente, pero muestran pequeñas variaciones debidas al muestreo aleatorio.

**¿Es lógico el resultado?**

Sí, ya que el embedding promedio representa una especie de "dígito prototípico" y la desviación estándar captura la variación entre las diferentes formas del dígito.

Al muestrear desde una N(μ,σ), estamos generando nuevas versiones plausibles del dígito dentro del espacio aprendido por el autoencoder.

**Conclusión**

Los resultados obtenidos demuestran que el autoencoder ha aprendido una representación latente(el espacio de menor dimensión donde el encoder codifica la informacion esencial de una imagenes) continua y significativa del espacio de los dígitos manuscritos.

El decoder, al recibir muestras generadas desde una distribución normal construida con la media y desviación estándar de los embeddings, es capaz de reconstruir imágenes visualmente coherentes y reconocibles, incluso si esos puntos no fueron vistos durante el entrenamiento. Esto indica que el modelo ha aprendido a generalizar correctamente dentro del espacio latente, respetando la estructura semántica de los datos originales, es decir, es capaz de distinguir cuando se trata de un 0, un 1 o un 8, aprendiendo sus características comunes.
